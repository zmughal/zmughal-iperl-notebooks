# An IPerl helper for `perldoc` HTML

author: Zaki Mughal

date: 2015-02-11

This notebook demonstrates how to add a helper function that reads in the output of `perldoc` and then displays it as HTML within the notebook.

```perl
IPerl->helper( perldoc_html => sub {
    shift;
    my $html = `perldoc -ohtml Try::Tiny 2> /dev/null`;
    return IPerl->html( $html );
});

IPerl->perldoc_html('Try::Tiny');
```
